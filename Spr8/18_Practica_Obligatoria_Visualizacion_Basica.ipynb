### #1.2

Recrea la siguiente figura, para ello construye en apartados separados cada grfico y luego haz un apartado en el que se junten los cuatro. Las figuras son:
1. Las ciudades ms pobladas, con los numeros de las poblaciones (tendrs que usar texto y recordar que para centrarlos se necesita poner el argumento `ha` a "center"). (Colores: azul, gris, aceituna, marrn, con trasparencia)
2. Histograma del rea total ocupada por las ciudades. (50 bins, fuente = "Bradley Hand itc", tamao 18, color azul acero)
3. El scatter de rea de agua de un ciudad con su poblacion.
4. El boxplot de la altura de las ciudades.

```python
y = df.sort_values("population_total", ascending=False)["population_total"].head(5)
```

```python
i_cities = df["population_total"].reset_index(drop=True).sort_values(ascending=False).head(5).index.to_list()
```

```python
x = df.iloc[i_cities]["city"]
```

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 8))

# Primer grfico (Barras)
axs[0, 0].bar(x, y / 1000, color=["blue", "grey", "green", "brown", "orange"], alpha=0.5)
axs[0, 0].set_title("Ciudades ms pobladas de California")
axs[0, 0].set_ylabel("Habitantes (miles)")
axs[0, 0].set_xlabel("Ciudades")
for i, value in enumerate(y):
    axs[0, 0].text(i, value / 1000 + 100, f"{value:,}", ha="center", fontsize=10)

# Segundo grfico (Histograma)
axs[0, 1].hist(df["area_land_km2"], bins=50, color="steelblue", alpha=0.7, label="Hist Area")
axs[0, 1].set_ylabel("Nmero de ciudades", fontdict={'family': 'Bradley Hand ITC', 'size': 12})
axs[0, 1].set_xlabel("Superficie ($\\mathregular{km^2}$)", fontdict={'family': 'Bradley Hand ITC', 'size': 12})
axs[0, 1].legend()

# Tercer grfico (Dispersin)
axs[1, 0].scatter(df["area_water_km2"], df["population_total"] / 1000, color="green", alpha=0.5)
axs[1, 0].set_ylabel("rea de agua de la ciudad ($\\mathregular{km^2}$)")
axs[1, 0].set_xlabel("Poblacin (miles habitantes)")
axs[1, 0].set_title("Poblacin por kilmetro cuadrado de agua")

outliers = df[df["population_total"] > 1000000] 
for idx, row in outliers.iterrows():
    x_outlier = row["area_water_km2"]
    y_outlier = row["population_total"] / 1000
    
    axs[1, 0].annotate(
        f"Outlier",  
        xy=(x_outlier, y_outlier),  
        xytext=(x_outlier - 50, y_outlier - 500), 
        arrowprops=dict(arrowstyle="->", lw=1.5), 
        fontsize=10,
        ha="left"
    )

# Cuarto grfico (Diagrama de caja)
box = axs[1, 1].boxplot(df["elevation_m"], vert=False, patch_artist=True)
for patch in box['boxes']:
    patch.set_facecolor('red')
axs[1, 1].set_xlabel("Altitud (m)")
axs[1, 1].set_title("Diagrama de caja de la Altitud")

# Ajustar espacio entre grficos
fig.tight_layout()
```
